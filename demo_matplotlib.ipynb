```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Line plot
plt.figure(figsize=(16, 12))
plt.subplot(2, 3, 1)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o')
plt.title('1. Line Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
```

```python
# 2. Bar plot
plt.subplot(2, 3, 2)
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 35]
plt.bar(categories, values, color='skyblue')
plt.title('2. Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# 3. Histogram
plt.subplot(2, 3, 3)
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, alpha=0.7, color='green')
plt.title('3. Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
# 4. Scatter plot
plt.subplot(2, 3, 4)
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x, y, alpha=0.6, color='red')
plt.title('4. Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')

# 5. Pie chart
plt.subplot(2, 3, 5)
sizes = [30, 25, 20, 25]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('5. Pie Chart')
```

```python
# 6. Box plot
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 6)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('6. Box Plot')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

```python
# 7. Subplots with different styles
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

# Line plot with multiple lines
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.plot(x, np.cos(x), label='cos(x)')
#ax1.plot(x, np.tan(x), label='tan(x)')
ax1.set_title('7. Multiple Lines')
ax1.legend()
ax1.grid(True)

# Area plot
ax2.fill_between(x, 0, np.sin(x), alpha=0.5, color='orange')
ax2.set_title('8. Area Plot')
ax2.grid(True)

# Stem plot
ax3.stem([1, 2, 3, 4, 5], [10, 25, 30, 20, 15])
ax3.set_title('9. Stem Plot')
ax3.grid(True)

# Error bar plot
x_err = [1, 2, 3, 4, 5]
y_err = [20, 25, 30, 35, 40]
yerr = [2, 3, 1, 4, 2]
ax4.errorbar(x_err, y_err, yerr=yerr, fmt='o', capsize=5)
ax4.set_title('10. Error Bar Plot')
ax4.grid(True)

plt.tight_layout()
plt.show()
```
