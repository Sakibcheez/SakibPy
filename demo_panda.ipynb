```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 #1. Creating DataFrames and Series
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'London', 'Tokyo', 'Paris']}
df = pd.DataFrame(data)
series = pd.Series([1, 2, 3, 4, 5])
print("1. Creating DataFrame and Series:")
print("DataFrame:\n", df)
print("\nSeries:", series.values)

# 2. Basic DataFrame info
print("\n2. DataFrame info:")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data types:\n", df.dtypes)

# 3. Selecting data
print("\n3. Selecting data:")
print("Single column (Name):", df['Name'].tolist())
print("Multiple columns:\n", df[['Name', 'Age']])
print("First row:\n", df.iloc[0])

# 4. Filtering data
print("\n4. Filtering data:")
adults = df[df['Age'] > 28]
print("People older than 28:\n", adults)

# 5. Adding and dropping columns
df['Salary'] = [50000, 60000, 70000, 55000]
print("\n5. Adding column:")
print(df)

df_dropped = df.drop('City', axis=1)
print("\nAfter dropping 'City' column:\n", df_dropped)
```

```python
# 6. Statistical summary
numeric_data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
print("\n6. Statistical summary:")
print(numeric_data.describe())

# 7. Grouping and aggregation
sales_data = pd.DataFrame({
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 120, 180, 160],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South']
})
grouped = sales_data.groupby('Product')['Sales'].sum()
print("\n7. Grouping and aggregation:")
print("Sales by Product:\n", grouped)

# 8. Sorting data
sorted_df = df.sort_values('Age', ascending=False)
print("\n8. Sorting by Age (descending):")
print(sorted_df)

# 9. Handling missing data
df_with_na = pd.DataFrame({
    'A': [1, 2, None, 4],
    'B': [5, None, 7, 8],
    'C': [9, 10, 11, None]
})
print("\n9. Handling missing data:")
print("Original with NaN:\n", df_with_na)
print("After dropping NaN:\n", df_with_na.dropna())
print("After filling NaN with 0:\n", df_with_na.fillna(0))

# 10. Reading/writing data (simulated)
print("\n10. Data I/O operations:")
# Create sample CSV data
sample_data = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Value': [10, 20, 30, 40]
})
print("Sample data for CSV operations:\n", sample_data)
# In real usage: sample_data.to_csv('data.csv', index=False)
# In real usage: loaded_data = pd.read_csv('data.csv')
```
